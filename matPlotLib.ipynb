## Mathematical Plotting Library : MatPlotLib 


```python
import matplotlib.pyplot as plt
import numpy as np

xPoint = np.array([0, 6])
yPoint = np.array([0, 200])

plt.plot(xPoint, yPoint, marker='o')
plt.show()
```

```python
plt.plot([1, 20, 15, 20], 'o:g')
plt.show()
```

```python
plt.plot([1, 20, 4, 2, 3, 90],'o-.r', ms=5)
plt.show()
```

### Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([1, 3, 2, 4, 5, 6])
y1 = np.array([3, 4, 5, 5, 9, 2])

plt.subplot(2, 1, 1) # For Column Type (1, 2, 1) 
plt.plot(x1, y1)

plt.subplot(2, 1, 2,)
plt.grid()
plt.plot(x1, y1, 'o-.r')

plt.show()
```

```python
plt.subplot(2, 3, 1)
plt.plot(x1, y1, marker='*')
plt.title('Sales', loc='left')

plt.subplot(2, 3, 2)
plt.plot(x1, y1, 'o-r')
plt.title('Market Cap')

plt.subplot(2, 3, 3)
plt.plot(x1, y1)
plt.title('Rewards', loc='Right')

plt.subplot(2, 3, 4)
plt.grid()
plt.plot(x1, y1, 'o-.g')

plt.subplot(2, 3, 5,)
plt.plot(x1, y1, '*-b') # marker|line|color

plt.subplot(2, 3, 6)
plt.plot(x1, y1)

plt.suptitle("Revenue!")
plt.show()
```

### Graphs

```python
# Scatter

x1 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y1 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

plt.scatter(x1, y1, color='g')

x2 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y2 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x2, y2, c='red')

x2 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y2 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x2, y2, c='red')

plt.show()
```

```python
x2 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y2 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])


c = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x2, y2, cmap='viridis', c=c, s=c, alpha=.8) # s(sizes of dots), c(colors), alpha(opectiy)
# plt.colorbar() # Optional

plt.show()

```

```python
x = np.random.randint(100, size=50)
y = np.random.randint(100, size=50)
s = 10 * np.random.randint(100, size=50)
c = np.random.randint(100, size=50)

plt.scatter(x, y, cmap='nipy_spectral',c=c, s=s, alpha=.7)
plt.colorbar()

plt.show()
```

```python
# Bar

x = np.random.randint(10, size=7)
y = np.random.randint(10, size=7)

plt.subplot(2, 1, 1)
plt.bar(x, y, color='g')


plt.subplot(2, 1, 2)
plt.barh(x, y, color='r')


plt.show()
```

```python
plt.bar(x, y, color='blue', width=1.3)
plt.show()
```

```python
# Histogram

x = np.random.randint(50, size=10)
plt.hist(x, color='grey')
plt.show()
```

```python
# Pie Chart
explode = [0, 0.2, 0, 0, 0]
x = np.array([10, 20, 40, 15, 5])

plt.pie(x, labels=x, explode=explode, startangle=90)
plt.legend(title='MyLabels')
plt.show()
```
