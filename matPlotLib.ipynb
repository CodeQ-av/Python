## Mathematical Plotting Library : MatPlotLib 


```python
import matplotlib.pyplot as plt
import numpy as np

xPoint = np.array([0, 6])
yPoint = np.array([0, 200])

plt.plot(xPoint, yPoint, marker='o')
plt.show()
```

```python
plt.plot([1, 20, 15, 20], 'o:g')
plt.show()
```

```python
plt.plot([1, 20, 4, 2, 3, 90],'o-.r', ms=5)
plt.show()
```

```python
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([8, 3, 2, 4, 5])
x2 = np.array([8, 3, 2, 4, 5])
y2 = np.array([8, 3, 2, 4, 5])

plt.plot(x1, y1, x2, y2)
plt.show()
```

```python

plt.title('Line Plot')
plt.grid(color='green', linestyle='--', linewidth=.3)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(xPoint, 'o-.r')
plt.show()
```
